```python
def calaculate_accuracy(x):
    X=df[x]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30)
    lgt_R = LogisticRegression(solver='lbfgs')
    lgt_R.fit(X_train, y_train)
    accuracy=lgt_R.score(X_test,y_test)
    print('Accuracy:',accuracy)
    return accuracy
```

```python
def normalize(chromosomes2):
    for l in range(len(chromosomes2)):
        for m in range(len(chromosomes2[l])):
            if chromosomes2[l][m]>0.5:
                chromosomes2[l][m]=1
            else:
                chromosomes2[l][m]=0
    return chromosomes2
```

```python
import numpy as np

class Particle:
    def __init__(self, num_dimensions, lower_bound, upper_bound):
        self.position = np.random.uniform(lower_bound, upper_bound, num_dimensions)
        self.velocity = np.random.uniform(-1, 1, num_dimensions)
        self.best_position = self.position
        self.best_fitness = float('inf')

def objective_function(x):
    # Define your objective function here
    # Example:
    return sum(x**2)

def particle_swarm_optimization(num_particles, num_dimensions, lower_bound, upper_bound, max_iterations, inertia_weight, cognitive_weight, social_weight):
    swarm = [Particle(num_dimensions, lower_bound, upper_bound) for _ in range(num_particles)]
    global_best_position = None
    global_best_fitness = float('inf')

    for _ in range(max_iterations):
        for particle in swarm:
            fitness = objective_function(particle.position)
            if fitness < particle.best_fitness:
                particle.best_position = particle.position
                particle.best_fitness = fitness
            if fitness < global_best_fitness:
                global_best_position = particle.position
                global_best_fitness = fitness

        for particle in swarm:
            r1 = np.random.rand(num_dimensions)
            r2 = np.random.rand(num_dimensions)
            particle.velocity = (inertia_weight * particle.velocity) + \
                                (cognitive_weight * r1 * (particle.best_position - particle.position)) + \
                                (social_weight * r2 * (global_best_position - particle.position))
            particle.position += particle.velocity

            # Boundary enforcement
            particle.position = np.clip(particle.position, lower_bound, upper_bound)

    return global_best_position, global_best_fitness

# Hyperparameters
num_particles = 20
num_dimensions = 2
lower_bound = -5.0
upper_bound = 5.0
max_iterations = 100
inertia_weight = 0.7
cognitive_weight = 1.5
social_weight = 1.5

best_position, best_fitness = particle_swarm_optimization(num_particles, num_dimensions, lower_bound, upper_bound, max_iterations, inertia_weight, cognitive_weight, social_weight)
print("Best position:", best_position)
print("Best fitness:", best_fitness)
```
